# Análise Exploratória em Conjunto de Dados do Kaggle

## Análise 3

```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
fig, ax = plt.subplots(figsize=(8, 5))
sns.barplot(x="fuelType", y="price", hue="gearbox", palette="husl", data=df)
ax.set_title("Preço médio dos veículos por tipo de combustível e tipo de caixa de câmbio", fontdict={'size':15})
ax.xaxis.set_label_text("Tipo de Combustível", fontdict={'size':15})
ax.yaxis.set_label_text("Preço Médio", fontdict={'size':15})
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price.png", facecolor="white")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
fig, ax = plt.subplots(figsize=(10, 6))
colors = ["windows blue", "amber", "greyish", "faded green", "dusty purple"]
sns.barplot(
  x="vehicleType", 
  y="powerPS", 
  hue="gearbox", 
  palette=sns.xkcd_palette(colors), 
  data=df
)
ax.set_title("Potência média dos veículos por tipo de veículo e tipo de caixa de câmbio", fontdict={'size':15})
ax.xaxis.set_label_text("Tipo de Veículo", fontdict={'size':15})
ax.yaxis.set_label_text("Potência Média", fontdict={'size':15})
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/vehicletype-fueltype-power.png", facecolor="white")
```
